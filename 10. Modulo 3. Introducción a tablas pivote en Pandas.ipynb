# 2 pivot_tables()

### tolera funciones aggregate

```python
# read and show the dataset
df_sales=pd.read_excel("supermarket_sales.xlsx")
df_sales
```

```python
# make a pivot table and add an aggregate function- cuando es sum valores, no se trabaja con texto
#solo con data num√©rica, por eso muestran las columnas con valores num√©ricos. Me aparecen todas las
#columnas
df_sales.pivot_table(index="Gender", aggfunc="sum")
```

```python
# make a pivot table, add and aggregate function and select some columns. Para eso agrego values=
#y entre corchetes y dentro de los corchetes, entre comillas, las columnas que le agrego
df_sales.pivot_table(index="Gender", values=["Quantity", "Total"], aggfunc="sum")
```

```python
#quiero probar si aparecen columnas con descripcion de texto aparecen siempre y cuando las ponga
#en el argumento columnas, enargunemto values no van a aparecer porque tienen que ser nros
df_sales.pivot_table(index="Gender", columns= "Payment", values=["Quantity", "Total"], aggfunc="sum")
```

```python
# make a pivot table that says how much male and female spend in each product line
#para eso agrego el argumento columnas, y pongo el nombre de la colunma y la subdivision interna
#que tenga me la va a mostrar
df_sales.pivot_table(index="Gender", columns="Product line", values="Total", aggfunc="sum")
```

```python
df_sales.pivot_table(index="Gender", columns="Product line", values=["Total", "Quantity"], aggfunc="sum")
```

```python
df_sales.pivot_table(index="Gender", columns="Payment", values=["Total", "Quantity"], aggfunc="sum")
```
