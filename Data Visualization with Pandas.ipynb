### New Dataset + Pivot Table

```python
import pandas as pd
```

```python
#leo archivo csv y le asigno una variable
df_population_raw = pd.read_csv("population_total.csv")
```

```python
#first look at the dataset. le pusieron raw por que tenemos que hacer unos cambios todavía
# para reshape, remodelar transform the dataframe
df_population_raw 
```

#                        1 Making a pivot table

### .pivot() Devuelve un reformateo de la DataFrame organizado por un index/valores en columna dado (pivote sin aggregates) usando pivot method

```python
#dropping null values. drop es dejar caer
#con este codigo ahora la tabla no deberia tener ninguna row null. de hecho antes habia 4185
# y ahora 4178, es decir 7 menos
df_population_raw.dropna
```

```python
df_population_raw.dropna(inplace=True)
```

```python
# para sobreescribir el contenido dentro de esta dataframe, reescribo la variable basicamente
df_population_raw = df_population_raw.dropna()
```

```python
df_population_raw.dropna()
```

```python
#making a pivot table
df_population_raw.pivot(index= "year", columns="country",  values="population")
```

```python
# le doy un nuevo nombre a mi dataframe con pivot table
df_population_raw.pivot(index= "year", columns="country",  values="population")
```

```python
df_pivot=df_population_raw.pivot(index= "year", columns="country",  values="population")
```

```python
# selecting some countries. selecciono solo algunas columnas/paises Y LO SOBREESCRIBO
df_pivot = df_pivot [["United States", "India", "China", "Indonesia", "Brazil", "Argentina"]]
```

```python
# showing pivot table
df_pivot
```

# 2 Lineplot, visualizacion con Pandas

```python
#import matplotlib
import matplotlib.pyplot as plt
```

```python
df_pivot.plot(kind="line")
```

```python
# para customizar más, puedo agregar otros argumentos, x label, etiqueta de eje x y etiqueta de 
# eje y (aparece escrito vertical) y título de toda la dataframe

df_pivot.plot(kind="line", xlabel="Year", ylabel="Population", title="Population (1959-2022)" )
```

```python
#podemos customizar el tamaño de la tabla con otro argumento. primero va el tamaño del eje x 
# y despues el tamaño del eje y, entre paréntesis y dividido por coma
df_pivot.plot(kind="line", xlabel="Year", ylabel="Population", title="Population (1959-2022)", 
              figsize=(8,4) )
import matplotlib.pyplot as plt
#save plot y entre paréntesis coloco el nombre que le quiero poner al archivo entre " .png"
plt.savefig("exportando_gráfico_de_lineas.pdf")
#show plot
plt.show()
```

```python
df_pivot.to_excel("pivot_table.xlsx")
```

# Piecharts

```python
#changing column name (column 2020) para eso usamos .rename
df_pivot_2020=df_pivot_2020.T
```

```python
df_pivot_2020
```

```python
# pasamos la columna 2020 integer a 2020 string. verde es integer, rojo es string
#inplace sirva para: para guardar estos cambios
#voy a utilizar inplace=True
df_pivot_2020.rename( columns={2020:"2020"}, inplace=True)
```

```python
#making the plotA
df_pivot_2020.plot(kind="pie", y="2020", title="Population 2020 (%)")
```

```python
df_pivot_2020.plot(kind="pie", y="2020", title="Population 2020 (%)", figsize=(6,6))
import matplotlib.pyplot as plt
#save plot y entre paréntesis coloco el nombre que le quiero poner al archivo entre " .png"
plt.savefig("piechart_torta.pdf")
#show plot
plt.show()
```
